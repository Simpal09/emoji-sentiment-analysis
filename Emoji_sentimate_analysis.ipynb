```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df= pd.read_csv('emoji.dataset.csv')

print(df.head())
print(df.info())
```

```python
print(df.describe())
```

```python
# frequency counts for categorical columns
print(df['Emoji'].value_counts())
print(df['Context'].value_counts())
print(df['Platform'].value_counts())
print(df['User Gender'].value_counts())
```

```python
# distribution of User Age
plt.figure(figsize=(10, 6))
sns.histplot(df['User Age'], bins=20, kde=True)
plt.title('Distribution of User Age')
plt.xlabel('User Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# emoji usage by Platform
plt.figure(figsize=(12, 8))
sns.countplot(y='Platform', data=df, order=df['Platform'].value_counts().index)
plt.title('Emoji Usage by Platform')
plt.xlabel('Count')
plt.ylabel('Platform')
plt.show()
```

```python
# emoji usage by User Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='User Gender', data=df)
plt.title('Emoji Usage by User Gender')
plt.xlabel('User Gender')
plt.ylabel('Count')
plt.show()
```

```python
# context associated with different Emojis
plt.figure(figsize=(14, 10))
sns.countplot(y='Context', data=df, order=df['Context'].value_counts().index)
plt.title('Context Associated with Different Emojis')
plt.xlabel('Count')
plt.ylabel('Context')
plt.show()
```

```python
# relationship between User Age and Emoji usage
plt.figure(figsize=(12, 8))
sns.boxplot(x='Emoji', y='User Age', data=df)
plt.title('User Age Distribution by Emoji')
plt.xlabel('Emoji')
plt.ylabel('User Age')
plt.xticks(rotation=90)
plt.show()
```

```python
#emoji usage across different Platforms
plt.figure(figsize=(14, 10))
sns.countplot(y='Emoji', data=df, hue='Platform', order=df['Emoji'].value_counts().index)
plt.title('Emoji Usage across Different Platforms')
plt.xlabel('Count')
plt.ylabel('Emoji')
plt.legend(title='Platform')
plt.show()
```

```python
sns.pairplot(df, hue='Context', palette='viridis')
plt.show()
```

```python
# violin plot of User Age by Context
plt.figure(figsize=(14, 10))
sns.violinplot(x='Context', y='User Age', data=df, palette='viridis')
plt.title('User Age Distribution by Context')
plt.xlabel('Context')
plt.ylabel('User Age')
plt.show()

# box plot of User Age by Platform
plt.figure(figsize=(14, 10))
sns.boxplot(x='Platform', y='User Age', data=df, palette='viridis')
plt.title('User Age Distribution by Platform')
plt.xlabel('Platform')
plt.ylabel('User Age')
plt.show()
```

```python
# ensure all Context values are strings and fill missing values with empty strings
df['Context'] = df['Context'].astype(str).fillna('')
```

```python
from textblob import TextBlob

# function to get the sentiment
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

df['Sentiment'] = df['Context'].apply(get_sentiment)
print(df[['Context', 'Sentiment']].head())

# descriptive statistics for sentiment scores
print(df['Sentiment'].describe())

# visualize sentiment distribution
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['Sentiment'], bins=20, kde=True)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
from wordcloud import WordCloud

# Generate a word cloud for all contexts
all_contexts = ' '.join(df['Context'])
wordcloud_all = WordCloud(width=800, height=400, background_color='white').generate(all_contexts)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud_all, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of All Contexts')
plt.show()

# Generate word clouds for specific contexts
contexts = df['Context'].unique()
for context in contexts:
    context_text = ' '.join(df[df['Context'] == context]['Context'])
    wordcloud_context = WordCloud(width=800, height=400, background_color='white').generate(context_text)
    
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud_context, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud of Context: {context}')
    plt.show()
```
